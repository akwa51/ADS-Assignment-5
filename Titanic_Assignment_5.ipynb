```python
import pandas as pd
import plotly.graph_objects as go
import plotly.express as px
import plotly.figure_factory as ff
```

```python
df=pd.read_csv('/content/gdrive/My Drive/Colab Notebooks/ADS-Assignment-5/Titanic Data.csv')
df.head()
```

```python
print(df.isnull().sum())
```

```python
df['Age'].fillna(round(df['Age'].mean(),2),inplace=True)
```

####Q1. Age Distribution Graph

```python
#Q1 Histogram Distribution
fig=px.histogram(df, x='Age', nbins=40,title='Titanic Age Distribution')
fig.show()
```

####Q2. Age Distribution Based on Gender

```python
fig=px.histogram(df,x='Age',title='Age Distribution By Gender', barmode='group',nbins=40,color='Sex')
fig.show()
```

####Q3.Matrix Plot of DataFrame

```python
fig = ff.create_scatterplotmatrix(df.corr(), diag='box',colormap='Portland',
                                  colormap_type='cat',
                                  height=700, width=700)
fig.show()
```

####Q4. Box Plot Representation of Fare and Pclass

```python
df_class=df.pivot(columns='Pclass', values='Fare')
df_class
```

```python
trace0 = go.Box(
    y=df_class[1],
    name = 'Class One',
    marker = dict(
        color = 'rgb(12, 12, 140)',
    )
)

trace1 = go.Box(
    y=df_class[2],
    name = 'Class Two',
    marker = dict(
        color = 'rgb(12, 128, 128)',
    )
)

trace2 = go.Box(
    y=df_class[3],
    name = 'Class Three',
    marker = dict(
        color = 'rgb(204, 0, 204)',
    )
)
data = [trace0, trace1, trace2]

layout = dict(title = 'Passenger Class Box Plot',
              xaxis=dict( title= 'Passenger Class'),yaxis= dict(title= 'Passenger Fare')
             )

fig = go.Figure(data=data,layout=layout)
fig.show()
```
